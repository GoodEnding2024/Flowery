```python
import random

def make_poem(keyword1, keyword2):
    
    k = keyword2[-1]
    
    josa = '을' if (ord(k)-44032)%28 else '를'

    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role":"user", "content": f"{keyword1}, {keyword2}{josa} 주제로 공백 포함 40글자 이내 짧은 시"}
        ]
    )
    
    return completion.choices[0].message.content

carnation = ['부모님의 사랑', '감동', '존경', '카네이션', '건강함']
poem_list = []

for _ in range(50):
    fl1, fl2 = random.sample(carnation, 2)
    poem = make_poem(fl1, fl2)
    
    print(len(poem))
    
    poem_list.append(poem)
```

```python
import re

new_poem_list = []

for i in range(len(poem_list)):
    poem = poem_list[i]
    
    pattern = re.compile(r'^[\'\"]|[\'\"]$')
    poem = re.sub(pattern, '', poem)
    
    pattern = re.compile(r'[a-zA-Z]')
    if bool(re.search(pattern, poem)):
        continue

    new_poem_list.append(poem)

poem_list = new_poem_list
```
